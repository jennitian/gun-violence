```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part one into a DataFrame.
map_df = pd.read_csv("Resources/gmap.csv")
map_df = map_df.dropna()
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
info_box_template = """
<dl>
<dt>Number of People Killed</dt><dd>{n_killed}</dd>
<dt>Notes</dt><dd>{notes}</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
shooting_info = [info_box_template.format(**row) for index, row in map_df.iterrows()]

locations = map_df[["latitude", "longitude"]]
# Get the maximum temperature.
n_killed = map_df["n_killed"]
# Assign the figure variable.
fig = gmaps.figure()
# Assign the heatmap variable.
heat_layer = gmaps.heatmap_layer(locations, weights=n_killed)
# Call the figure to plot the data.
marker_layer = gmaps.marker_layer(locations, info_box_content=shooting_info)
fig.add_layer(heat_layer)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
